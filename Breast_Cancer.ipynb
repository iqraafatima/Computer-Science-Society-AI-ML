```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)

x_pca = np.concatenate((x_train_pca, x_test_pca), axis=0)
y_combined = pd.concat([y_train, y_test], axis=0).reset_index(drop=True)

df_pca = pd.DataFrame(data=x_pca, columns=['principal_component_1', 'principal_component_2'])
df_pca['diagnosis'] = y_combined

display(df_pca.head())
```
